```python
import pandas as pd
import numpy as np
import re
```

```python
RAW_XLS = '../國防部/102年度1-10月份依預算法第62條之1辦理政策宣導相關之廣告調查表-3.xls'
```

```python
df = pd.read_excel(RAW_XLS, 0, skiprows=2, skipfooter=2)
```

```python
df.head()
```

```python
# 可能的時間格式
# Case A: 102年2月6日, 102年2月
get_date_A = re.compile(
    '(?P<year>\d{3})年'
    '(?P<mo>\d+)月'
    '((?P<d>\d+)日|)',
    re.UNICODE
).search

# Case B: 102年7月1日起至7月31日
get_date_B = re.compile(
    '(?P<year>\d{3})年'
    '(?P<s_mo>\d+)月(?P<s_d>\d+)日'
    '起?至'
    '((?P<e_mo>\d+)月|)'
    '(?P<e_d>\d+)日',
    re.UNICODE
).search


# Case C:
# 102.10.1-10.31
# 1020926-1021025
# 1020926~1021025
get_date_C = re.compile(
    '(?P<year>\d{3})[.]?'
    '(?P<s_mo>\d{1,2})[.]?'
    '(?P<s_d>\d{1,2})'
    '[-~]'
    '\d{3}[.]?'
    '((?P<e_mo>\d{1,2})[.]?|)'
    '(?P<e_d>\d{1,2})',
    re.UNICODE
).search

# 102.08.28
# 1020926
# 102/1/20
get_date_D = re.compile(
    '(?P<year>\d{3})[./]'
    '(?P<mo>\d{1,2})[./]'
    '(?P<d>\d{1,2})',
    re.UNICODE
).search

get_date_E = re.compile(
    '(?P<year>\d{3})'
    '(?P<mo>\d{2})'
    '(?P<d>\d{2})',
    re.UNICODE
).search

# 7/18
# 5月份
get_date_F = re.compile('(?P<mo>\d{1,2})/(?P<d>\d{1,2})', re.UNICODE).search
get_date_G = re.compile('(?P<mo>\d+)月[份號]?((?P<d>\d{1,2})日|)', re.UNICODE).search

# 全年度

search_year = re.compile('(?P<year>\d{3})年', re.UNICODE).search

def find_year_in_title(title):
    m = search_year(title)
    if m:
        return m.groupdict()['year']
    else:
        return None

def parse_single_date(raw_title, raw_d):
    m = get_date_B(raw_d) or get_date_C(raw_d)
    if m:
        d = m.groupdict()
        date = '{}/{}-{}/{}'.format(
            int(d['s_mo']), int(d['s_d']),
            (int(d['e_mo']) if d['e_mo'] else int(d['s_mo'])),
            int(d['e_d'])
        )
        return d['year'], date 
    
    m = get_date_A(raw_d) or get_date_D(raw_d) or get_date_E(raw_d)
    if m:
        d = m.groupdict()
        date = ('{}/{}'.format(int(d['mo']), int(d['d'])) if d['d'] else int(d['mo']))
        return d['year'], '{}-{}'.format(date, date)
    
    m = get_date_F(raw_d) or get_date_G(raw_d)
    if m:
        year = find_year_in_title(raw_title) or ''
        d = m.groupdict()
        if d['d']:
            date = '{}/{}'.format(d['mo'], d['d'])
        else:
            date = '{}'.format(d['mo'])
        return year, '{}-{}'.format(date, date)
    
    return None

def parse_time(row):
    proj, title, raw_dates = row[[2, 3, 5]]
    ret_val = [
        parse_single_date(proj+title, raw_d) for raw_d in raw_dates.splitlines()
    ]
    # print(ret_val)
    y = ', '.join(set(v[0] for v in ret_val if v))
    d = ', '.join([v[1] for v in ret_val if v])
    return (y or np.NaN), (d or np.NaN)
```

```python
df_time = pd.DataFrame([parse_time(r) for _, r in df.iterrows()], columns=['民國年', '時間'])
df = pd.concat([df, df_time], axis=1)
# df[pd.isnull(df_time.iloc[:,0]) & pd.isnull(df_time.iloc[:,1])].head()
```

```python
note_tpl = '''\
宣導計畫、政策內容：{0[2]}
媒體執行細節：{0[5]}
預算類別：{0[8]}
'''

def make_note(row):
    return note_tpl.format(row)

df['備註'] = df.apply(make_note, axis=1)
```

```python
new_index = df.columns.values
new_index[1] = '執行單位'
new_index[3] = '主題'
new_index[4] = '媒體'
new_index[10] = '金額'
df.columns = new_index
df['行政部會'] = '國防部'
df = df[['主題', '媒體', '民國年', '時間', '金額', '執行單位', '行政部會', '備註']]
```

```python
df.loc[df.loc[:, '金額'].isnull(), '金額'] = 0
```

```python
df = df.ffill()
```
