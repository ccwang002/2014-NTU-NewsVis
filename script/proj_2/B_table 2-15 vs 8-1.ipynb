```python
import numpy as np
from pathlib import Path
import pandas as pd
import re

ATTACH_ROOT = Path("../../2_職災分析/附件/")
OUT_ROOT = ATTACH_ROOT.parent / '圖表輸出'
TAB_ROOT = ATTACH_ROOT.parent / '整理好的單表'
```

## 讀表 2-15

```python
attach_5 = str(ATTACH_ROOT / "5.xls")
```

```python
df_215 = pd.read_excel(attach_5, 1, skiprows=2, skip_footer=3)
```

```python
second_col = df_215.irow(0).values
from_first_col = pd.isnull(second_col)
second_col[from_first_col] = df_215.columns[from_first_col]
new_col = pd.Series([re.sub(r'[\n ]+', '', cell, re.MULTILINE&re.UNICODE) for cell in second_col])
new_col[:3] = ['行業別', '實施檢查廠次', '違反法令項數']
# new_col.head(10)
```

```python
new_col = pd.Series(
    [re.sub(r'[\n ]+', '', cell, re.MULTILINE & re.UNICODE)
     for cell in df_215.irow(0).astype(str)],
    dtype=str
)
new_col.iloc[:3] = ['行業別', '實施檢查廠次', '違反法令項數']
# new_col.head(10)
```

```python
top_col = pd.Series(
    [re.sub(r'[\n ]+', '', cell, re.MULTILINE & re.UNICODE)
     for cell in df_215.columns],
    dtype=str
)
process_ix = ~top_col.apply(lambda c: "行業別" in c)
data_col = new_col.loc[process_ix]
data_col.loc[data_col == 'nan'] = np.NaN
data_col.ffill(inplace=True)
new_col.loc[process_ix] = data_col[process_ix]
# new_col.head(10)
```

```python
# 增加廠次、%、項數 
tp_list = (new_col.loc[process_ix]).loc[3:].iteritems()
for (i1, c1), (i2, c2), (i3, c3) in zip(*[iter(tp_list)]*3):
    new_col[i1] += '廠次'
    new_col[i2] += '%'
    new_col[i3] += '項數'

df_215.columns = new_col
left_ix = [not ("行業別" in c) for c in df_215.columns]
left_ix[0] = True
df_215 = df_215.loc[:, left_ix]

df_215 = pd.DataFrame(df_215.iloc[2:, df_215.columns != 'nan'], copy=True)
```

```python
# 去掉行業別空白、多餘的裝飾
def clean_jobtype(jobtype):
    return re.sub('[ *]+', '', jobtype, re.UNICODE)

df_215.loc[:, "行業別"] = df_215["行業別"].apply(clean_jobtype)

# 去掉總計的資料行
ix = ~df_215['行業別'].isin(["總計", "製造業"])
df_215 = df_215.ix[ix, :]

# 使用行業別作為 index
df_215.index = df_215.iloc[:, 0]
df_215 = df_215.iloc[:, 1:]
```
