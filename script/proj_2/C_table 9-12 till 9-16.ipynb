```python
import numpy as np
from pathlib import Path
import pandas as pd
import re

ATTACH_ROOT = Path("../../2_職災分析/附件/")
OUT_ROOT = ATTACH_ROOT.parent / '圖表輸出'
TAB_ROOT = ATTACH_ROOT.parent / '整理好的單表'
```

```python
attach_13 = str(ATTACH_ROOT / "13.xls")
```

```python
insur_rawdata = [pd.read_excel(attach_13, i, skiprows=2, skip_footer=3, na_values=['-']) for i in range(3, 8)]
```

```python
def my_div(col_num, col_denom):
    mod_denom = col_denom.copy()
    mod_denom[mod_denom == 0] = np.NAN
    return col_num / mod_denom

def compute_item_avg(df_raw):
    # only extract what we want
    df = pd.DataFrame(df_raw.iloc[1:, 3:].copy())
    df.index = df_raw.iloc[1:, 0].values
    # give full column names
    new_col = [col.replace(' ', '') for c in df.columns[::2] for col in [c+'件數', c+'金額']]
    df.columns = new_col
    # compute avg money per item
    df_avg = pd.DataFrame({money[:-2] + '每件金額': my_div(df[money], df[item]) for item, money in zip(*[iter(df.columns)]*2)})
    return df_avg
```

```python
insur_avg = [compute_item_avg(df).fillna(value=1) for df in insur_rawdata]
```

```python
titles = [
    '全產業',
    '農、林、漁、牧',
    '礦業及土石採取業',
    '製造業',
    '營造業'
]
```

```python
C_ROOT = OUT_ROOT / "C"
if not C_ROOT.exists():
    C_ROOT.mkdir()
    
for title, df in zip(titles, insur_avg):
    df.to_excel(
        str(C_ROOT / 'C_{}每件平均給付金額.xlsx'.format(title))
    )
```

```python
def plot_avg(df, title):
    fig, ax = plt.subplots()
    fig.set_size_inches(10, 5)
    n_grp = len(df)
    bar_width = 0.9
    opacity = 0.9
    index_pos = 4.5 * np.arange(n_grp) + 1

    for i, col in enumerate(df.columns):
        plt.bar(index_pos + bar_width * i, df[col], bar_width,
                log=True,
                alpha=opacity,
                color = color_set.hex_colors[i],
                label = col[:-4])

    plt.xticks(index_pos + bar_width * 2, [ix[2:-1] for ix in df.index])
    plt.ylim(10**2, 2*10**6)
    plt.legend(prop=chtfont, loc='center left', bbox_to_anchor=(1, 0.75))
    ax.grid(False, axis='x')
    plt.xlabel("民國年", fontproperties=chtfont)
    plt.ylabel("金額（元）", fontproperties=chtfont)
    plt.title(title + "平均每件給付金額", fontproperties=chtfont)
    plt.savefig(str(C_ROOT / "C_{}每件平均給付金額(對數).svg".format(title)), bbox_inches='tight')
```

```python
for df, title in zip(insur_avg, titles):
    plot_avg(df, title)
```
